### HRMS DATAFRAME CREATING WITH PYTHON ###

```python
import pandas as pd
import numpy as np
from faker import Faker
from warnings import filterwarnings
filterwarnings('ignore')

# Importing faker for creating a fake data which used for creating data frame 
fake = Faker()

# using faker seed is creating a fake data frame with constant repeated output 
Faker.seed(0)

# Create an empty list to hold the employee data
employee_data = []

# Generate 12 HR employees
for i in range(12):
    employee = {
        "EmployeeID": fake.unique.random_int(min=1000, max=1012),
        "FirstName": fake.first_name(),
        "LastName": fake.last_name(),
        "Gender": fake.random_element(elements=("Male", "Female")),
        "Age": fake.random_int(min=23, max=65),
        "Department": "HR",
        "Position": fake.job(),
        "DateOfHire": fake.date_between(start_date="-10y", end_date="today"),
        "Salary": round(fake.random_number(digits=5), -2),  
        "PerformanceScore": fake.random_int(min=1, max=5),
        "Report_admin": 1000
    }
    employee_data.append(employee)

# Generate 88 non-HR employees
for j in range(188):
    employee = {
        "EmployeeID": fake.unique.random_int(min=1000, max=9999),
        "FirstName": fake.first_name(),
        "LastName": fake.last_name(),
        "Gender": fake.random_element(elements=("Male", "Female")),
        "Age": fake.random_int(min=21, max=65),
        "Department": fake.random_element(elements=("Engineering", "Marketing", "Sales", "Finance")),
        "Position": fake.job(),
        "DateOfHire": fake.date_between(start_date="-10y", end_date="today"),
        "Salary": round(fake.random_number(digits=5), -2),  # Salary rounded to nearest hundred
        "PerformanceScore": fake.random_int(min=1, max=5),
        "Report_admin": fake.random_int(min=1001, max=1012)
    }
    employee_data.append(employee)


# Create the DataFrame
df = pd.DataFrame(employee_data)

# Display the entire DataFrame
print(df)

pd.set_option('display.max_rows', None)
```

```python
# creating a visualization for the hrms project data 

import matplotlib.pyplot as plt
import seaborn as sns

# Remove unwanted warnings

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# Visualizations
plt.figure(figsize=(12, 8))

# Bar Chart: Number of employees in each department
plt.subplot(2, 2, 1)
bars=sns.countplot(data=df, x='Department')
plt.title('Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Count')


# Histogram: Distribution of ages of employees
plt.subplot(2, 2, 2)
sns.histplot(df['Age'], bins=10, kde=True)
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Count')

# Box Plot: Salary distribution by department
plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='Department', y='Salary')
plt.title('Salary Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Salary')

# Count Plot: Gender distribution across departments
plt.subplot(2, 2, 4)
sns.countplot(data=df, x='Department', hue='Gender')
plt.title('Gender Distribution Across Departments')
plt.xlabel('Department')
plt.ylabel('Count')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
